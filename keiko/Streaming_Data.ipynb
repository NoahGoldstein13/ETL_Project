```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import csv
streaming_file = 'data/MoviesOnStreamingPlatforms_updated.csv'
streaming_data_df = pd.read_csv(streaming_file)
streaming_data_df.sample(10)
```

```python
#picking columns and assigning to new dataframe
streaming_cols = ['Title', 'Age','IMDb', 'Rotten Tomatoes', 'Netflix', 'Hulu', 
                  'Prime Video', 'Disney+', 'Directors', 'Genres', 'Country', 
                  'Language', 'Runtime']

streaming_cleaned = streaming_data_df[streaming_cols]

streaming_cleaned
```

```python
#adding a calculated field 



streaming_cleaned['on_platforms_besides_netflix'] = streaming_cleaned['Hulu'] + streaming_cleaned['Prime Video'] + streaming_cleaned['Disney+']


streaming_cleaned.sample(50)
```

```python
#duplicating new title column for joining with other datasets
streaming_cleaned['title_for_join']= streaming_cleaned['Title']


streaming_cleaned
```

```python
#stripping spaces and making them lower case
streaming_cleaned['title_for_join'] = streaming_cleaned['title_for_join'].str.replace(' ', '')
streaming_cleaned['title_for_join'] = streaming_cleaned['title_for_join'].str.lower()

streaming_cleaned.sample(10)
```

```python
#stripping additional characters to match title 
chars = "\\`*_{}[]()>#+-.!$:;%&,'/?"
for c in chars:
    streaming_cleaned['title_for_join'] = streaming_cleaned['title_for_join'].str.replace(c,"")
streaming_cleaned.sample(100)
```

```python
#dropping duplicates 
final_streaming_df = streaming_cleaned.drop_duplicates()
```
