```python
import pandas as pd
```

```python
netflix_dataset = "../datasets/netflix_titles.csv"
netflix_df = pd.read_csv(netflix_dataset)
netflix_df.head()
```

```python
new_netflix_df = netflix_df[['title','cast', 'date_added','release_year', 'rating', 'description']]

# Rename the column headers
new_netflix_df = new_netflix_df.rename(columns={"title": "netflix_title",
                                                "date_added": "date_added_to_netflix",
                                                "rating":"maturity_rating"})

new_netflix_df
```

```python
new_netflix_df['title_for_join']=new_netflix_df['netflix_title'].str.strip()
```

```python
new_netflix_df['title_for_join']=new_netflix_df['title_for_join'].str.replace(" ",'')
new_netflix_df['title_for_join']=new_netflix_df['title_for_join'].str.lower()
new_netflix_df.sample(10)
```

```python
chars = "\`*_{}[]()>#+-.,!$:;%'&/?"
for c in chars:
    new_netflix_df['title_for_join'] = new_netflix_df['title_for_join'].str.replace(c,"")

new_netflix_df
```

```python
clean_netflix_df = new_netflix_df.drop_duplicates()
clean_netflix_df.head(20)
```

```python
clean_netflix_df.to_csv(r'../clean_datasets/clean_netflix_data.csv', index = False)
```
