```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
streaming_data = "clean_datasets/clean_streaming.csv"
streaming_df = pd.read_csv(streaming_data)
streaming_df.head()
```

```python
netflix_data = "clean_datasets/clean_netflix_data.csv"
netflix_df = pd.read_csv(netflix_data)
netflix_df.head()
```

```python
tmdb_data = "clean_datasets/clean_TMDb_data.csv"
tmdb_df = pd.read_csv(tmdb_data)
tmdb_df.head()
```

```python
# Create a filtered dataframe from specific columns
streaming_cols = ["Title", "Age", "IMDb", "Rotten Tomatoes", "Directors","Genres", "Country", "Language", "Runtime", "on_platforms_besides_netflix", "title_for_join"]
streaming_cols_chart = streaming_df[streaming_cols]

# Rename the column headers
streaming_cols_chart = streaming_cols_chart.rename(columns={"Rotten Tomatoes": "rotten_tomatoes",
                                                           "Title": "title",
                                                           "Age": "age", 
                                                            "IMDb": "imdb", 
                                                            "Directors": "director",
                                                            "Genres": "genres", 
                                                            "Country": "country", 
                                                            "Language": "language",
                                                            "Runtime": "runtime"})


streaming_cols_chart.head()
```

```python
#Rename Column 

netflix_df = netflix_df.rename(columns={"cast": "movie_cast"})

netflix_df.head()
```

```python
#connect engine

connection_string = "postgres:postgres@localhost:5433/netflix_data"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
netflix_df.to_sql(name='netflix_data', con=engine, if_exists='append', index=False)
```

```python
streaming_cols_chart.to_sql(name='streaming_data', con=engine, if_exists='append', index=False)
```

```python
tmdb_df.to_sql(name='tmdb_data', con=engine, if_exists='append', index=False)
```
